```python
import pandas as pd
import uuid
import random
from faker import Faker
```

```python
num_users = 100
```

```python
features = [
    "id",
    "gender",
    "subscriber",
    "name",
    "rating"
]

df = pd.DataFrame(columns=features)
df
```

```python
df['id'] = [uuid.uuid4().int for i in range(num_users)]
df
```

```python
print(df['id'].nunique()==num_users)
```

```python
genders = ["male","female","na"]

df['genders'] = random.choices(
    genders,
    weights=(45,45,10),
    k=num_users
)
df
```

```python
choice = [True, False]

df['subsricber'] = random.choices(
    choice,
    k=num_users
)
df
```

```python
faker = Faker(locale="EN_IN")
def name_gen(gender):


    if gender=='male':
      return faker.name_male()
    elif gender=='female':
      return faker.name_female()

    return faker.name()

df['name'] = [name_gen(i) for i in df['gender']]
df
```

```python
genders = ["male","female","na"]

df['genders'] = random.choices(
    genders,
    weights=(45,45,10),
    k=num_users
)
df
```

```python
choice = ["1","2","3","4","5"]
df['rating'] = random.choices(
    choice,
    weights=(45,45,10,42,60),
    k=num_users
)
df
```

```python
df.to_csv('dataset1.csv')
```
